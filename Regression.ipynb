```python
data = house_data[house_data['arrondissement'] > 0]
```

```python
X = np.matrix([np.ones(data.shape[0]), data['arrondissement'].as_matrix(), 
               data['surface'].as_matrix()]).T
y = np.matrix(data['price']).T
```

```python
theta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
print(theta)
```
